```python
a = df.iloc[:, 1:51]
m = min(a.iloc[0,:])
bg = a.iloc[0,:]-m
df.iloc[0, 1:51] = bg
df.iloc[0, 1:51]
```

```python
col_count = [50, 50, 50, 50, 50, 50, 21, 50, 50, 44]

col_start = 1
col_end = 1
nrows = len(df)
for x in range(len(col_count)):
    col_end += col_count[x]
    for y in range(nrows):
        df_part = df.iloc[:, col_start:col_end]
        row_min = min(df_part.iloc[y,:])
        row_bg_rem = df_part.iloc[y,:]-row_min
        df.iloc[y, col_start:col_end] = row_bg_rem
    col_start += col_count[x]
```
