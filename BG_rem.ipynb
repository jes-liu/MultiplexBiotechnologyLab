```python
# subtracting by min value per batch per row
col_count = [50, 50, 50, 50, 50, 50, 21, 50, 50, 44]

col_start = 1
col_end = 1
nrows = len(df)
for x in range(len(col_count)):
    col_end += col_count[x]
    for y in range(nrows):
        df_part = df.iloc[:, col_start:col_end]
        row_min = min(df_part.iloc[y,:])
        row_bg_rem = df_part.iloc[y,:]-row_min
        df.iloc[y, col_start:col_end] = row_bg_rem
    col_start += col_count[x]
```

```python
# subtracting by harmonic avg value per batch per row

def harmean(nums):
    # nums = list of values
    n = len(nums)
    
    sub = [0] * n
    for i in range(n):
        sub[i] = 1/nums[i]
    
    mean = n/sum(sub)
    
    return mean

col_count = [50, 50, 50, 50, 50, 50, 21, 50, 50, 44]

col_start = 1
col_end = 1
nrows = len(df)
for x in range(len(col_count)):
    col_end += col_count[x]
    for y in range(nrows):
        df_part = df.iloc[:, col_start:col_end]
        bg_value = harmean(df_part.iloc[y,:])
        row_bg_rem = df_part.iloc[y,:]-bg_value
        for z in range(len(row_bg_rem)):
            if row_bg_rem[z] < 0:
                row_bg_rem[z] = 0
        #print(bg_value)
        #print(row_bg_rem)
        df.iloc[y, col_start:col_end] = row_bg_rem
    col_start += col_count[x]
```

```python
dir = 'C:/Users/jesse/OneDrive/Documents/Multiplex Lab/Data/Pseudo-cells_preprocess.xlsx'

df_old = pd.read_excel(dir)
```

```python
len(df_old.columns)-1
```

```python
nrow = len(df_old)
ncol = len(df_old.columns)-1
df = df_old.iloc[:, 1:(ncol+1)]
df
```

```python
for x in range(ncol):
    bg = np.percentile(df.iloc[:, x], 75)
    for y in range(nrow):
        old_num = df.iloc[y, x]
        new_num = old_num - bg
        if new_num < 0:
            df.iloc[y, x] = 0
        else:
            df.iloc[y, x] = new_num

df_bg = df
df_bg
```

```python
drop = []
for x in range(nrow):
    if df_bg.iloc[x, :].any() == 0:
        drop.append(x)
df_drop = df_bg.drop(drop)
len(drop)
```

```python
from sklearn.preprocessing import normalize

colnames = df_drop.columns

# log normalize per cell
df_cellnorm = pd.DataFrame(normalize(df_drop, axis=1), columns=colnames)
# log2(x+1)
df_celllognorm = np.log2(df_cellnorm+1)
```

```python
nrow_new = nrow-len(drop)

for x in range(nrow_new):
    for y in range(ncol):
        if df_celllognorm.iloc[x, y] < .5:
            df_celllognorm.iloc[x, y] = 0
        else:
            pass
```
